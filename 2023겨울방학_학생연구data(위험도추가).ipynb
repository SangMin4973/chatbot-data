```python
df_1.set_index('회사', inplace=True)

# 각 행과 모든 컬럼에 대한 값을 출력
answer = []
for row_index, row in df_1.iterrows():
    for column, value in row.items():
       result = f"'{row_index}'의 '{column}'은(는) {value}이다."
       answer.append(result)
```

```python
import re
import csv

question_data_list = []
answer_data_list = []

for item in answer:
    date_match = re.search(r'\d{4}/\d{2}', item)
    if date_match:
        date = date_match.group()
        formatted_date = f"{date[:4]}년 {int(date[5:]):02d}월"
    else:
        formatted_date = "N/A"

    company_name = re.sub(r'\(주\)', '', item.split("'")[1].split(',')[1].strip())
    value = item.split()[-1]
    attribute_name = item.split("'")[3]

    que_data = f"{company_name}의 {formatted_date} {attribute_name}은 얼마야"
    ans_data = f"(주){company_name}의 {formatted_date} {attribute_name}은 {value.replace('이다.', ' 입니다.')}"

    question_data_list.append(que_data)
    answer_data_list.append(ans_data)
```

```python
answer_data_list
```

```python
import csv

# CSV 파일로 저장
file_path_1 = 'answer.csv'
with open(file_path_1, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(answer_data_list)

file_path_2= 'question.csv'
with open(file_path_2, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(question_data_list)

print(f'CSV 파일이 성공적으로 저장되었습니다. 경로: {file_path_1} , {file_path_2}')
```
